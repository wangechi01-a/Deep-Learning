## Types of Activation function implemented in python code:
- > Tanh
- > Sigmoid
- > ReLU
- > Leaky ReLU

```python
import math

def sigmoid(x):
    return 1/(1+math.exp(-x))
```

```python
sigmoid(100)
```

```python
sigmoid(2)
```

```python
sigmoid(-20)
```

```python
sigmoid(0.8)
```

```python
def tanh(x):
  return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))
    
```

## Tanh

```python
tanh(-50)
```

```python
tanh(1)
```

## ReLU

```python
def relu(x):
    return max(0,x)
```

```python
relu(-70)
```

```python
relu(5)
```

## Leaky ReLU

```python
def leaky_relu(x):
    return max(0.1*x,x)
```

```python
leaky_relu(-30)
```

```python
leaky_relu(4)# For positive values it does not make a change
```
